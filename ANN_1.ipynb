```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-10,10,100)
y=1/(1+np.exp(-x))
plt.plot(x,y)
plt.title("sigmoid")
plt.xlabel("input")
plt.ylabel("activation")
plt.show()
```

```python
x=np.linspace(-10,10,100)
y=np.tanh(x)
plt.plot(x,y)
plt.title("tanh")
plt.xlabel("input")
plt.ylabel("activation")
plt.show()
```

```python
x=np.linspace(-10,10,100)
y=np.maximum(0,x)
plt.plot(x,y)
plt.title("relu")
plt.xlabel("input")
plt.ylabel("activation")
plt.show()
```

```python
x=np.linspace(-10,10,100)
alpha=0.5
y=np.maximum(x,alpha*x)
plt.plot(x,y)
plt.title("leaky relu")
plt.xlabel("input")
plt.ylabel("activation")
plt.show()
```

```python
x=np.linspace(-10,10,100)
y=np.exp(x)/np.sum(np.exp(x))
plt.plot(x,y)
plt.title("softmax")
plt.xlabel("input")
plt.ylabel("activation")
plt.show()
```

```python
def mp(x1,x2):
    w1=1
    w2=-1
    thre=0.5
    wei_sum=x1*w1+x2*w2
    if wei_sum>thre:
        return 1
    else:
        return 0
print(mp(0,0))
print(mp(1,0))
print(mp(0,1))
print(mp(1,1))
```
